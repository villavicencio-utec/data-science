```python
#Drop missing value

df.dropna(subset = ["symboling"], axis = 0, inplace = True)
#axis 0 is for row and 1 is for columns
#inplace True able to modify the df and if we use False , we have to assign to a new df.
```

```python
#Replace value

df_mean = df[df["price"]!="?"]
df_mean["price"] = df_mean["price"].astype("int")
mean_1 = df_mean["price"].mean()
mean = int(mean_1)
mean
```

```python
df["price"] = df["price"].replace('?', mean)
```

```python
df["city-mpg"] = 235/df["city-mpg"]
```

```python
df.rename(columns = {"city-L/100Km" : "city-L/100km"}, inplace = True)
```

```python
df.columns
```

```python
for i in range(len(df["price"])):    
    print(df["price"][i])
    
```

```python
check_nan = df['price'].isnull().values.any()
print(check_nan)
```

```python
df['price'] = df['price'].astype(int) 
```

```python
df.dtypes
```

```python

df
```

```python
#Simple feature scaling
df["symboling"] = df["symboling"]/df["symboling"].max()
```

```python
df
```

```python
#Min-Max
df["price"] = (df["price"] - df["price"].min())/(df["price"].max() - df["price"].min())
```

```python
df
```

```python
df["city-L/100km"].plot()
```

```python
df_aux = df["city-L/100km"]
df_aux
```

```python
df_aux.plot()
```

```python
#Z-core
df_aux_z_core = (df_aux - df_aux.mean())/(df_aux.std())
df_aux_z_core.plot()
# df["city-L/100km"] = (df["city-L/100km"] - df["city-L/100km"].mean())/(df["city-L/100km"].std())
```

```python
#Min-Max
df_aux_min_max = (df_aux - df_aux.min())/(df_aux.max() - df_aux.min())
df_aux_min_max.plot()
```

```python
#Simple feature scaling
df_aux_simple_scaling = df_aux/df_aux.max()
df_aux_simple_scaling.plot()
```
